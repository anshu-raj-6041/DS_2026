#### Conditional Statements

```python
## if statement

age=2

if age>=18:
    print("U can vote")
else:
    print("U can't vote")
```

```python
## elif

age=15

if age<13:
    print("U are a child")
elif age<18:
    print("U are a teenager")
else:
    print("U are adult")
```

```python
## Nested Conditional Statements
## no is even, odd or negative

num = int(input("Enter the no:"))

if num>=0:
    print(f"{num} is +")
    if num%2==0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
else:
    print(f"{num} is -")
```

```python
## Practical Ex
## Leap year or not using nested if-else condition

year = int(input("Enter the year:"))

if year%4==0:
    if year%100==0:
        if year%400==0:
            print(f"{year} is a leap year")
        else:
            print(f"{year} is not a leap year")
    else:
        print(f"{year} is a leap year")
else:
    print(f"{year} is not a leap year")
```

```python
## Assignment
## Simple Calculator

n1 = float(input("Enter the first no:"))
n2 = float(input("Enter the second no:"))
op = input("Choose the operator :: +, -, *, /")

if op=='+':
    res=(n1+n2)
elif op=='-':
    res=(n1-n2)
elif op=='*':
    res=(n1*n2)
elif op=='/':
    if n2!=0:
        res=(n1/n2)
    else:
        print("Divison by 0, error")
else:
    res="Invalid Input"

print("Result =>", res)
```
