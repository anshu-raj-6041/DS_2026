#### Loops

```python
range(5)
```

```python
## for loop

for i in range(5):
    print(i)
```

```python
## for loop

for i in range(1,10):
    print(i)
```

```python
range(1,10,2)
```

```python
for i in range(1,10,2):
    print(i)
```

```python
## for loop

for i in range(10,1,-1):
    print(i)
```

```python
## strings :: collection of characters

str="Anshu Raj"

for i in str:
    print(i)
```

```python
## while loop
## while loop continues to execute as long as the condition is true

count=0

while count<5:
    print(count)
    count=count+1
```
