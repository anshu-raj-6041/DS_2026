#### Operators

```python
## Arithmetic Operation

a=10
b=5

add_res=a+b
sub_res=a-b
mult_res=a*b
div_res=a/b
floor_res=a//b
modulus_res=a%b
exponential_res=a**b

print(add_res)
print(sub_res)
print(mult_res)
print(div_res)      ##gives ans in float
print(floor_res)    ##gives ans in int (lower value deta hai)
print(modulus_res)  ##gives remainder
print(exponential_res)
```
