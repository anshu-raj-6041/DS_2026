```python
## Understanding Variables types
## Python is dynamically typed, type of a var is determined at runtime

age=25  ##int
height=6.1  ##float
name="Anshu"    ##string
is_student=False    ##bool

print(type(age))
print(type(height))
print(type(name))
print(type(is_student))
```

```python
## Dynamic Typing
## Python allows the type of a var to change as the program executes

var=10
print(var, type(var))

var="India"
print(var, type(var))

var=10.5
print(var, type(var))
```

```python
## simple calculator

n1=float(input("Enter the 1st no:"))
n2=float(input("Enter the 2nd no:"))

sum=n1+n2
diff=n1-n2
product=n1*n2
quotient=n1/n2

print("Sum =>", sum)
print("Difference =>", diff)
print("Product =>", product)
print("Divison =>", quotient)
```
